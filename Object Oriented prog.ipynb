### Object Oriented

The main idea when using the formal object oriented method is to create figure objects and then just call methods or attributes of that object. This approach is nicer when dealing with a canvas that has multiple plots on it

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
x = np.linspace(0,20,10)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
#  create figure = empty canvas
fig = plt.figure()

# add set of axes to the figure

axes = fig.add_axes([0.1,0.1,0.8,0.8])



axes.plot(x,y,"r-")
axes.set_xlabel("X Axis")
axes.set_ylabel("Y Axis")
axes.set_title("Linear Graph")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title("MAIN PLOT")
axes2.set_title("SUB PLOT")
```

```python
fig, axes= plt.subplots(nrows= 1, ncols =2)
# iterating through an axes array as with a normal array
for ax in axes:
    ax.plot(x,y)
```
